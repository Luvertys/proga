```python
import requests
from PIL import Image
from IPython.display import display
import io

api_key = 'zeA2hRvUOaAOAXCDAFFU4eeSL8j2pvWq0AxHeguD'

def APOD():
    url = f"https://api.nasa.gov/planetary/apod?api_key={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        print("\nАстрономическая картинка дня: ")
        print(f"дата: {data['date']}")
        print(f"название: {data['title']}")
        print(f"описание: {data['explanation']}")
        print(f"ссылка на изображение: {data['url']}\n")
        img_response = requests.get(data['url'])
        img = Image.open(io.BytesIO(img_response.content))
        display(img)
    else:
        print("ошибка при загрузке данных")

def pictures_mars(data, camera):
    url = f"https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos"
    params = {
        'earth_date': data,
        'camera': camera,
        'api_key': api_key
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data['photos']:
            for photo in data['photos']:
                print(f"фото URL: {photo['img_src']}")

                img_response = requests.get(photo['img_src'])
                img = Image.open(io.BytesIO(img_response.content))
                display(img)
        else:
            print("нет фоток")
    else:
        print("ошибка при загрузке данных")

def OSZ(date):
    url = f"https://api.nasa.gov/neo/rest/v1/feed"
    params = {
        'start_date': date,
        'end_date': date,
        'api_key': api_key
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        nos = data['near_earth_objects'].get(date, [])
        print(f"\nОбъекты, сближающиеся с Землей {date}: ")
        for i in nos:
            print(f"название: {i['name']}, диаметр (км): {i['estimated_diameter']['kilometers']['estimated_diameter_max']}, потенциально опасен: {i['is_potentially_hazardous_asteroid']}")
    else:
        print("ошибка при загрузке данных")

def space_weather():
    url = f"https://api.nasa.gov/DONKI/notifications?api_key={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        events = response.json()
        if events:
            print("\nданные о космической погоде:")
            for event in events:
                print(f"событие: {event['messageType']}, дата: {event['messageIssueTime']}, подробности: {event['messageBody']}")
        else:
            print("нет данных")
    else:
        print("ошибка при загрузке данных")

def run():
    while True:
        print("\nфункционал")
        print("1. Астрономическая картинка дня (APOD)")
        print("2. Фотографии с марсохода")
        print("3. Объекты, сближающиеся с Землей (ОСЗ)")
        print("4. Данные о космической погоде (оно работает, но лучше не жмать)")
        print("5. Выйти")

        choice = input("выберите действие: ")

        if choice == '1':
            APOD()
        elif choice == '2':
            date = input("введите дату (гггг-мм-дд): ")
            camera = input("введите тип камеры (FHAZ): ")
            pictures_mars(date, camera)
        elif choice == '3':
            date = input("введите дату (гггг-мм-дд): ")
            OSZ(date)
        elif choice == '4':
            space_weather()
        elif choice == '5':
            print("спарта")
            break
        else:
            print("елзе")

run()
```

Разрешите пользователям перемещаться по страницам с произведениями искусства, выполнять фильтрацию по исполнителю или выходить из программы.

Если страниц с произведениями искусства несколько, укажите варианты перехода к следующей странице, предыдущей странице, фильтрации по исполнителю или выхода из программы.

```python
import requests
from PIL import Image
from IPython.display import display
import io

url = "https://api.artic.edu/api/v1/artworks"


def get_artworks(page=1, limit=10):
    params = {
        "page": page,
        "limit": limit,
        "fields": "id,title,artist_title,date_display,medium_display"
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        return data['data'], data['pagination']
    else:
        print("ошибка при загрузке данных")
        return [], {}


def filter_by_artist(artworks, artist_name):
    artist_name = artist_name.lower()
    filtered = [art for art in artworks if artist_name in (art.get('artist_title') or "").lower()]
    return filtered


def artwork_inf(artwork):
    print(f"\nназвание: {artwork['title']}")
    print(f"художник: {artwork.get('artist_title', 'ноунейм')}")
    print(f"дата: {artwork.get('date_display', 'неизвестно')}")
    print(f"материал: {artwork.get('medium_display', 'неизвестно')}\n")


def run():
    page = 1
    while True:
        artworks, pagination = get_artworks(page)

        if not artworks:
            break

        print(f"\nсписок произведений искусства (стр. {page}):")
        for i, artwork in enumerate(artworks, 1):
            print(f"{i}. {artwork['title']}")

        action = input("\nфункционал"
                       "\n1 посмотреть подробности произведения"
                       "\n2 фильтр по художнику"
                       "\n3 следующая страница"
                       "\n4 предыдущая страница"
                       "\n5 ввести нужную страницу"
                       "\n6 выход"
                       "\nвыберите действие: ")

        if action == "1":
            try:
                choice = int(input("\nвведите номер произведения: "))
                if 1 <= choice <= len(artworks):
                    artwork_inf(artworks[choice - 1])
                else:
                    print("неверный номер")
            except ValueError:
                print("неверный номер")

        elif action == "7459":
            break

        elif action == "2":
            artist_name = input("введите имя художника: ")
            filtered_artworks = filter_by_artist(artworks, artist_name)
            if filtered_artworks:
                print("\nпроизведения по запросу:")
                for artwork in filtered_artworks:
                    artwork_inf(artwork)
            else:
                print("ничего не найдено")

        elif action == "3":
            if pagination['current_page'] < pagination['total_pages']:
                page += 1
            else:
                print("больше только моя самооценка")

        elif action == "4":
            if page > 1:
                page -= 1
            else:
                print("назад пути нет")

        elif action == "5":
            inputed_str = int(input("введите страницу: "))
            if 0 < inputed_str < pagination['total_pages']:
                page = inputed_str
            else:
                print("разогнался")


        elif action == "6":
            image_url = "https://sun9-45.userapi.com/s/v1/ig2/cPatWbQSKKDCv8p3Bt3kHnPmesO0imJ8eOewMa427f78DrlAkCRT7s7tzRF_bEkChYG25JkFYiPPlCGOMe7w3u03.jpg?quality=96&as=32x32,48x48,72x72,108x108,160x160,240x240,360x360,480x480,540x540,640x640&from=bu&u=qZEpeZ3tQw8AzIBMaDVMRVwoF6Yac0v1oQhUOY3jw_8&cs=640x640"
            response = requests.get(image_url)
            if response.status_code == 200:
                img = Image.open(io.BytesIO(response.content))
                display(img)
            break
        else:
            print("неверный номер")


run()
```

# Задание 3

Задача: Создать программу по управлению портфелем криптовалют

Цель: Создать скрипт на Python, который извлекает цены на криптовалюты в режиме реального времени, позволяет пользователям управлять портфелем криптовалют, вычисляет общую стоимость портфеля, отслеживает изменения цен и предоставляет исторические данные о ценах для анализа.

Требования:
Получение текущих цен на криптовалюты:

Используйте https://docs.coingecko.com/ для получения актуальных цен на список криптовалют.

Управление портфелем:

- Позволяет пользователю создавать портфель криптовалют и управлять им, указывая количество каждой криптовалюты, которой он владеет.
- Расчитывает общую стоимость портфеля в указанной фиатной валюте (например, долларах США).

Отслеживание изменения цен:

- Отображение процентного изменения цены для каждой криптовалюты в портфеле за последние 24 часа.
- Выделите все криптовалюты, стоимость которых значительно увеличилась или снизилась.

Поиск исторических данных о ценах:

- Получение исторических данных о ценах на указанную криптовалюту за последнюю неделю.
- Предоставьте пользователю возможность визуализировать эти данные в простом текстовом формате (например, цены за день).

Взаимодействие с пользователем:

- Реализуйте интерфейс командной строки для ввода данных пользователем.
- Предоставьте опции для получения текущих цен, управления портфелем, просмотра изменений цен или анализа исторических данных.


```python
import requests
from datetime import datetime


class FutureMajor:
    def __init__(self):
        self.portfel = {}

    def get_coin_price(self, coin):
        url = "https://api.coingecko.com/api/v3/simple/price"
        params = {"ids": coin, "vs_currencies": "usd"}
        response = requests.get(url, params=params)

        if response.status_code == 200:
            data = response.json()
            if coin in data:
                return data[coin].get("usd", 0)
            else:
                print(f"не найдены данные для монеты {coin.upper()}")
        else:
            print(f"ошибка при загрузке данных")
        return 0

    def add_to_portfel(self, coin, pozitsiya):
        self.portfel[coin] = self.portfel.get(coin, 0) + pozitsiya
        print(f"{pozitsiya} {coin.upper()} добавлено в портфель")

    def get_portfel_value(self):
        total_value = 0
        for coin, pozitsiya in self.portfel.items():
            price = self.get_coin_price(coin)
            if price:
                value = price * pozitsiya
                total_value += value
                print(f"{coin.upper()}: {pozitsiya} x {price:.2f} USD = {value:.2f} USD")
        print(f"общая стоимость портфеля: {total_value:.2f} USD")
        return total_value

    def price_24timeframe(self):
        print("\nизменение цены за последние 24 часа: ")
        for coin in self.portfel.keys():
            url = f"https://api.coingecko.com/api/v3/coins/{coin}"
            response = requests.get(url)
            if response.status_code == 200:
                data = response.json()
                price_change = data.get("market_data", {}).get("price_change_percentage_24h", 0)
                print(f"{coin.upper()}: изменение на {price_change:.2f}%")
                if price_change > 20:
                    print(f"{coin.upper()} значительно увеличилась")
                elif price_change < -20:
                    print(f"{coin.upper()} значительно снизилась")
            else:
                print(f"ошибка при загрузке данных")

    def get_historical_data(self, coin):
        url = f"https://api.coingecko.com/api/v3/coins/{coin}/market_chart"
        params = {"vs_currency": "usd", "days": "7"}
        response = requests.get(url, params=params)
        if response.status_code == 200:
            data = response.json()
            prices = data.get("prices", [])
            dates = [datetime.fromtimestamp(price[0] / 1000).strftime('%Y-%m-%d') for price in prices]
            values = [price[1] for price in prices]
            print(f"\nисторические данные для {coin.upper()}: ")
            for date, value in zip(dates, values):
                print(f"{date}: {value:.2f} USD")
        else:
            print(f"ошибка при загрузке данных")



def run():
    diary_crypto_investor = FutureMajor()
    while True:
        print("\nфункционал")
        print("1. получить текущие цены криптовалют")
        print("2. добавить криптовалюту в портфель")
        print("3. рассчитать стоимость портфеля")
        print("4. показать изменения цен в портфеле")
        print("5. получить исторические данные о монете")
        print("6. выход")
        choice = input("выберите действие: ")

        if choice == "1":
            coin = input("введите название монеты: ").lower()
            price = diary_crypto_investor.get_coin_price(coin)
            if price:
                print(f"текущая цена {coin.upper()}: {price:.2f} USD")
        elif choice == "2":
            coin = input("введите название монеты: ").lower()
            quantity = float(input("введите количество: "))
            diary_crypto_investor.add_to_portfel(coin, quantity)
        elif choice == "3":
            diary_crypto_investor.get_portfel_value()
        elif choice == "4":
            diary_crypto_investor.price_24timeframe()
        elif choice == "5":
            coin = input("введите название монеты: ").lower()
            diary_crypto_investor.get_historical_data(coin)
        elif choice == "6":
            print("типичный мув хамстеркомбатовца")
            break
        else:
            print("мимо")


run()
```
